```python
import matplotlib.pyplot as plt

# Grouping by Date to sum doses for visualization
doses_per_day = df.groupby('Updated On')[['Male (Doses Administered)', 'Female (Doses Administered']].sum()

plt.figure(figsize=(12, 6))

# Plot for First Dose
plt.plot(doses_per_day.index, doses_per_day['Male (Doses Administered)'], label='First Dose (Male)', color='blue')

# Plot for Second Dose
plt.plot(doses_per_day.index, doses_per_day['Female (Doses Administered)'], label='Second Dose (Male)', color='green')

plt.title('First and Second Doses Administered to Males Over Time')
plt.xlabel('Updated On')
plt.ylabel('Number of Doses')
plt.legend()
plt.show()
```
