```python
import pandas as pd
```

```python
import seaborn as sns
```

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
df=sns.load_dataset('titanic')
```

visualization on the basis of two values
  - categorical data
  - numerical data

Basic Graph Types
   - LINE GRAPH: show trends over time
   - BAR GRAPH: compare values between categories(discrete data)
   - PIE CHART:compare values between categories (percentage data)
   - HISTOGRAM:show


```python
sns.countplot(df,x='class')
plt.title('Passengers count by class')
plt.xlabel("Class")
plt.ylabel("Count")
plt.show()
```

```python
##count plot:survival based on gender
##hue category btata hai
sns.countplot(df,x='survived',hue='sex',palette='RdBu')
plt.title('survived vs not survived by gender')
plt.show()
```

```python
sns.barplot(df,x='sex',y='fare',hue='class')
plt.title('Average Fare by Passenger Class and Gender')
```

```python
sns.pointplot(df,x='class',y='survived',hue='sex',markers=['o','x'])
plt.title('Survival Rate by Class and Gender')
plt.show()
```
